```python
import warnings
warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

## Data Preparation

```python
# read crypto_data.csv into Pandas
df = pd.read_csv("crypto_data.csv")
df
```

```python
# discard all cryptocurrencies that are not being traded
crypto_df = df.loc[(df["IsTrading"]==True),:]
```

```python
# drop the IsTrading column from the dataframe
crypto_df.drop(columns=["IsTrading","Unnamed: 0"], inplace=True)
```

```python
# drop rows that have at least one missing value
crypto_df.dropna(how="any", inplace=True)
```

```python
# Filter for cryptocurrencies where TotalCoinsMined is greater than 0
crypto_df = crypto_df.loc[(crypto_df["TotalCoinsMined"])>0]
crypto_df
```

```python
# Save targe column for later
labels =crypto_df["CoinName"]
labels
```

```python
# drop column=CoinName from dataframe
crypto_df.drop(columns=["CoinName"],inplace=True)
crypto_df.reset_index(drop=True, inplace=True)
crypto_df
```

```python
crypto_df.info()
```

```python
# convert data type of TotalCoinSupply to numeric
crypto_df["TotalCoinSupply"]=pd.to_numeric(crypto_df["TotalCoinSupply"])
```

```python
crypto_df.info()
```

```python
# Convert remaining features with text values into numerical data
df_encoded = pd.get_dummies(crypto_df)
df_encoded
```

```python
# Standardize your dataset so that columns that contain larger values do not unduly influence outcome
crypto_scaled = StandardScaler().fit_transform(df_encoded)
print(crypto_scaled)
```

```python
crypto_scaled.shape
```

## Dimensionality Reduction

```python
# Instantiate PCA model that will preserve approximately 99% of the explained variance
pca = PCA(n_components=0.90)
```

```python
crypto_pca = pca.fit_transform(crypto_scaled)
crypto_pca
```

```python
crypto_pca.shape
```

```python
# Initialize tsne model 
tsne = TSNE(learning_rate=100)
```

```python
# reduce dimensions
tsne_features = tsne.fit_transform(crypto_pca)
tsne_df = pd.DataFrame(data=tsne_features)
```

```python
plt.scatter(tsne_df[0],tsne_df[1],c=labels.index)
plt.show()
```

## Cluster Analysis with k-Means

```python
inertia = []
k=list(range(1, 10))

for i in k:
    km = KMeans(n_clusters=i)
    km.fit(tsne_df)
    inertia.append(km.inertia_)

elbow_data ={"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow
```

```python
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,10))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve for crypto')
plt.show()
```

## Cryptocurrencies can be clustered together because the plot of TSNE shows distinct 3-4 clusters, and the elbow chart has a distinct elbow at around 3 & 4 clusters.
